### Clean the text column Removing the Regexp

```python
def clean_tweet_text(text):
    """
    Comprehensive tweet cleaning function
    """
    if pd.isna(text):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+', '', text)
    
    # Remove user mentions and hashtags (but keep the text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#', '', text)
    
    # Remove special characters and numbers
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra whitespace
    text = ' '.join(text.split())
    
    return text
    # Apply cleaning
df['cleaned_text'] = df['text'].apply(clean_tweet_text)
```

### Tokenization and Lemmatization  
- Tokenization process of splitting text into smaller units called tokens.
-  Lemmatization reduces a word to its base or diction
- Processed text stopwords

```python

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words and len(token) > 2]
    return ' '.join(tokens)

df['processed_text'] = df['cleaned_text'].apply(preprocess_text)
```

## Dealing with Nan Value

```python

# 2. Détecter les marques dans la colonne tweet_text
def detect_brand_in_text(text):
    """
    Détecte si le texte mentionne Apple ou Google
    """
    if pd.isna(text):
        return 'unknown'
    
    text_str = str(text).lower()
    
    # Mots-clés pour Apple
    apple_keywords = ['apple', 'iphone', 'ipad', 'mac', 'ios', 'macbook', 'imessage', 'apple watch', 'app store']
    # Mots-clés pour Google
    google_keywords = ['google', 'android', 'pixel', 'nexus', 'android auto', 'google photos', 
                      'google assistant', 'play store', 'chrome', 'youtube']
    
    has_apple = any(keyword in text_str for keyword in apple_keywords)
    has_google = any(keyword in text_str for keyword in google_keywords)
    
    if has_apple and has_google:
        return 'both'
    elif has_apple:
        return 'apple'
    elif has_google:
        return 'google'
    else:
        return 'other'

# 3. Appliquer la détection sur tout le dataset
df['detected_brand'] = df['text'].apply(detect_brand_in_text)

print("\nMarques détectées dans le texte:")
print(df['detected_brand'].value_counts())

# 4. Remplir les NaN dans sentiment avec les marques détectées
def fill_missing_brands(row):
    """
    Remplit les valeurs NaN avec la marque détectée dans le texte
    """
    if pd.isna(row['sentiment']):
        return row['detected_brand']
    else:
        return row['sentiment']

# Appliquer la correction
df['sentiment'] = df.apply(fill_missing_brands, axis=1)

# 5. Nettoyer et standardiser les valeurs
df['sentiment'] = (
    df['sentiment']
    .astype(str)
    .str.lower()
    .str.strip()
    .replace('nan', 'unknown')  # Au cas où il reste des NaN convertis en string
)
```

```python
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
# Initialiser VADER
sia = SentimentIntensityAnalyzer()

def detect_sentiment(text):
    if not isinstance(text, str):  # Vérifie si ce n'est pas une chaîne
        text = str(text)           # Convertit en chaîne
    score = sia.polarity_scores(text)['compound']
    if score > 0.05:
        return "Positive"
    elif score < -0.05:
        return "Negative"
    else:
        return "Neutral"

# Appliquer à la colonne 'text'
df['sentiment2'] = df['text'].apply(detect_sentiment)
```

```python
import pandas as pd

def fill_brand_mentions(df, brand_col='detected_brand'):
    """
    Vérifie apple_mention et google_mention. Si NaN, les remplit selon detected_brand.
    """
    # Apple
    df['apple_mention'] = df['apple_mention'].fillna(False)  # Remplace NaN par False temporairement
    df.loc[df['apple_mention'] == False, 'apple_mention'] = df.loc[df['apple_mention'] == False, brand_col].str.contains(
        r'apple|iphone|ipad|mac', case=False, na=False
    )
    
    # Google
    df['google_mention'] = df['google_mention'].fillna(False)
    df.loc[df['google_mention'] == False, 'google_mention'] = df.loc[df['google_mention'] == False, brand_col].str.contains(
        r'google|android|pixel', case=False, na=False
    )
    
    return df


df = fill_brand_mentions(df)

df
```

### Step 5: Model Building
#### Prepare Data for Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

```

### Encode target variable

```python
le = LabelEncoder()
y = le.fit_transform(df_clean['sentiment'])
```

### Train-test split

```python

X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Training set: {X_train.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")
```

### Model Training & Evaluation

```python
# Initialize models
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'SVM': SVC(kernel='linear', random_state=42)
}

results = {}

for name, model in models.items():
    # Train model
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred = model.predict(X_test)
    
    # Evaluate
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = {
        'model': model,
        'accuracy': accuracy,
        'predictions': y_pred
    }
    
    print(f"{name} Accuracy: {accuracy:.4f}")
    print(classification_report(y_test, y_pred, target_names=le.classes_))
    print("="*50)
```

```python
best_model = results['Logistic Regression']['model']
best_name = max(results, key=lambda k: results[k]['accuracy'])
best_model = results[best_name]['model']
print(f"Best model: {best_name}")

```
